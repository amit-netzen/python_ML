```python
class shape :
    def __init__(self,name) :
        self.name = name
        def area(self) :
            pass    
class circle(shape) :
    def __init__(self,radius) :
        super().__init__("circle")
        self.radius = radius
    def area(self) :
        return 3.14 * self.radius * self.radius 
class rectangle(shape) :  
    def __init__(self,width,height) :
        super().__init__("rectangle")
        self.width = width
        self.height = height
    def area(self) :
        return self.width * self.height
c = circle(5)
print(f"The area of the circle is : {c.area()}")        
r = rectangle(4,5)
print(f"The area of the rectangle is : {r.area()}")
```

```python
def describe_shape(s) :
    print(f"The shape is : {s.name} and its area is : {s.area()}")  
describe_shape(c)
describe_shape(r)   
```

```python
from abc import ABC,abstractmethod
class vehicle(ABC) :
    def __init__(self,name,model) :
        self.name = name
        self.model = model
    @abstractmethod
    def fuel_type(self) :
        pass
class car(vehicle) :
    def __init__(self,name,model) :
        super().__init__(name,model)
    def fuel_type(self) :
        return "Petrol"
class bike(vehicle) :
    def __init__(self,name,model) :
        super().__init__(name,model)
    def fuel_type(self) :
        return "Diesel" 
v1 = car("Honda","City")
print(f"The fuel type of {v1.name} {v1.model} is : {v1.fuel_type()}")
v2 = bike("Yamaha","R15")
print(f"The fuel type of {v2.name} {v2.model} is : {v2.fuel_type()}")
```
