```python
import sqlite3 as sql
conn = sql.connect('test.db')
conn 
```

```python
cursor = conn.cursor()
cursor
```

```python

cursor.execute("""
    CREATE TABLE IF NOT EXISTS employee (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER NOT NULL,
    department TEXT NOT NULL,
    UNIQUE(name, age, department)
);
""")
conn.commit()
```

```python
cursor.execute('''INSERT INTO employee (name, age, department)
                VALUES (?, ?, ?)''', ("John Doe", 30, "HR"))
conn.commit()
```

```python
employees =[
    ("John Doe", 30, "HR"),
    ("Jane Smith", 25, "IT"),
    ("Bob Johnson", 35, "Finance"),
    ("Alice Brown", 28, "Marketing"),
    ("Charlie Wilson", 32, "IT")

]
cursor.executemany('''INSERT INTO employee (name, age, department)
                VALUES (?, ?, ?)''', employees)
conn.commit()
```
