```python
data=pd.DataFrame({'catogry':[np.random.choice(['A','B','C']) for _ in range(10)],
                   'value':[np.random.randint(1, 100) for _ in range(10)]})
print("Original DataFrame:\n", data)
data_grouped=data.groupby('catogry')
print("Grouped DataFrame:\n", data_grouped.sum())
print("Mean of each group:\n", data_grouped.mean())
```

```python
data=pd.DataFrame({'product':[np.random.choice(['A','B','C']) for _ in range(10)],
                   'category':[np.random.choice(['X','Y','Z']) for _ in range(10)],
                   'sales':[np.random.randint(1, 100) for _ in range(10)]})
print("Original DataFrame:\n", data)
grouped_data=data.groupby(['category'])
total_sales=grouped_data['sales'].sum()
print("Total sales for each category:\n", total_sales)
```

```python

import pandas as pd
import numpy as np
data1 = pd.DataFrame({
    'sale_date': [
        pd.Timestamp('2023-01-01') +
        pd.to_timedelta(np.random.randint(0, 365), unit='D')
        for _ in range(10)
    ],
    'category': [np.random.choice(['X','Y','Z']) for _ in range(10)],
    'price': [np.random.randint(1,100) for _ in range(10)]
})
print("Original DataFrame:\n", data1)
merged_data = data1.groupby([data1['sale_date'].dt.to_period('M'), 'category'])['price'].sum().reset_index()
print("Total sales for each month and category:\n", merged_data)
```
