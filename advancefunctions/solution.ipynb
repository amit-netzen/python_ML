```python
def feb(number, MEMO={}):
 
  if number == 0:
    return 0
  elif number == 1:
    return 1
  if number in MEMO:
    return MEMO[number]
  else:
    result = feb(number - 1, MEMO) + feb(number - 2, MEMO)
    MEMO[number] = result
    return result
print(feb(10))
```

```python
def fun(**kwargs):
    result = {}
    for key , value in kwargs.items():
       if isinstance(value, int):
           result[key] = value
    return result
print(fun(a=1, b=2, c='string', d=3.14))
```
